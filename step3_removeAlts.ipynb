```python
setChr=set(df['seqid'])
print(len(setChr))
```

```python
outFP=open("./hg38_p12_genbank_chromosomes_prim.fa","w+")

numChrFound=0
writeFlag=False

with open("/home/sparrow/genomicData/hg38/hg38_p12_genbank_chromosomes_all.fa",'r') as refFP:
    for line in refFP.readlines():
        if line[0]==">":
            chrID=line.split(" ")[0][1:].strip()
            if chrID in setChr:
                writeFlag=True
                outFP.write(line)
                numChrFound+=1
                print(numChrFound)
            else:
                writeFlag=False
        else:
            if writeFlag:
                outFP.write(line)

outFP.close()
```
