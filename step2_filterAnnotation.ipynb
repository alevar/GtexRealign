```python
import pandas as pd
import numpy as np
import subprocess
import csv
import os

gff3Cols=["seqid","source","type","start","end","score","strand","phase","attributes"]
```

```python
df=pd.read_csv("./hg38_p12_gencode29_refseq.gff3",sep='\t',names=gff3Cols)
df.head()
```

```python
dfG=df[df['type']=='gene'].reset_index(drop=True)
dfG
```

```python
l=['transcribed_unprocessed_pseudogene', 'unitary_pseudogene', 'rRNA_pseudogene', 'polymorphic_pseudogene', 'translated_processed_pseudogene', 'TR_V_pseudogene', 'processed_pseudogene', 'TR_J_pseudogene', 'IG_V_pseudogene', 'unprocessed_pseudogene', 'pseudogene', 'IG_C_pseudogene', 'transcribed_processed_pseudogene', 'transcribed_unitary_pseudogene', 'IG_pseudogene', 'IG_J_pseudogene']
dfG["gene_type"]=dfG['attributes'].str.split("gene_type=",expand=True)[1].str.split(";",expand=True)[0]
for i in l:
    print(i,set(dfG[dfG['gene_type']==i]["source"]))
```

```python
dfG["pseudo"]=dfG.attributes.str.split("pseudo=",expand=True)[1].str.split(";",expand=True)[0]
set(dfG[dfG["pseudo"]=="true"]["source"])
set(dfG["pseudo"])
```

```python
# one by one let's investigate how to remove pseudogenes from the annotation
dfG["pseudo"]=dfG.attributes.str.split("pseudo=",expand=True)[1].str.split(";",expand=True)[0]
dfG=dfG[~(dfG['pseudo']=="true")].reset_index(drop=True)[gff3Cols]
pseudogene_gene_type=[x for x in set(dfG.attributes.str.split("gene_type=",expand=True)[1].str.split(";",expand=True)[0]) if not x==None and "pseudo" in x]
print(pseudogene_gene_type)
dfG["gene_type"]=dfG['attributes'].str.split("gene_type=",expand=True)[1].str.split(";",expand=True)[0]
dfG=dfG[~(dfG["gene_type"].isin(pseudogene_gene_type))].reset_index(drop=True)
dfG
# how about transcribed/translated/processed pseudogenes
```

```python
dfG[dfG["attributes"].str.contains('pseudo')]
```

```python
dfG['geneID']=dfG.attributes.str.split("ID=",expand=True)[1].str.split(";",expand=True)[0]
setGeneIDs=set(dfG['geneID'])
dfG['transID']=dfG['geneID']
print(len(setGeneIDs))
```

```python
# now need ot use these geneIDs to select correct transcripts that correspond to the finalized set of genes
dfT=df[df["type"]=="transcript"].reset_index(drop=True)
dfT.head()
```

```python
dfT["pseudo"]=dfT.attributes.str.split("pseudo=",expand=True)[1].str.split(";",expand=True)[0]
dfT=dfT[~(dfT['pseudo']=="true")].reset_index(drop=True)[gff3Cols]
pseudogene_gene_type=[x for x in set(dfT.attributes.str.split("gene_type=",expand=True)[1].str.split(";",expand=True)[0]) if not x==None and "pseudo" in x]
print(pseudogene_gene_type)
dfT["gene_type"]=dfT['attributes'].str.split("gene_type=",expand=True)[1].str.split(";",expand=True)[0]
dfT=dfT[~(dfT["gene_type"].isin(pseudogene_gene_type))].reset_index(drop=True)
dfT
```

```python
pseudogene_transcript_type=[x for x in set(dfT.attributes.str.split("transcript_type=",expand=True)[1].str.split(";",expand=True)[0]) if not x==None and "pseudo" in x]
print(pseudogene_transcript_type)
dfT["transcript_type"]=dfT['attributes'].str.split("transcript_type=",expand=True)[1].str.split(";",expand=True)[0]
dfT=dfT[~(dfT["transcript_type"].isin(pseudogene_transcript_type))].reset_index(drop=True)
dfT
```

```python
dfT['geneID']=dfT.attributes.str.split("Parent=",expand=True)[1].str.split(";",expand=True)[0]
setTransParents=set(dfT["geneID"])
dfG=dfG[dfG['geneID'].isin(setTransParents)].reset_index(drop=True)
setGeneIDs=set(dfG['geneID'])
```

```python
dfT=dfT[dfT["geneID"].isin(setGeneIDs)].reset_index(drop=True)
print(len(dfT))
dfT['transID']=dfT.attributes.str.split('ID=',expand=True)[1].str.split(";",expand=True)[0]
dfT.head()
setTransIDs=set(dfT['transID'])
print(len(setTransIDs))
```

```python
# now to subset the exons:
dfE=df[df['type']=='exon'].reset_index(drop=True)
```

```python
dfE['transID']=dfE.attributes.str.split("Parent=",expand=True)[1].str.split(";",expand=True)[0]
print(len(dfE))
dfE=dfE[dfE["transID"].isin(setTransIDs)].reset_index(drop=True)
print(len(dfE))
dfE=dfE.merge(dfT[['transID','geneID']].drop_duplicates(keep='first'),how='inner',on='transID')
dfE.head()
```

```python
# now to join all the three features together
df_new=pd.concat([dfG,dfT,dfE],axis=0)
df_new
```

```python
df_new=df_new[gff3Cols+["geneID",'transID']].reset_index(drop=True)
df_new
```

```python
del setGeneIDs
del setTransIDs
del dfG
del dfT
del dfE
```

```python
df_new["type"]=pd.Categorical(df_new["type"],categories=["gene","transcript","exon"],ordered=True)
df_new=df_new.sort_values(by=["geneID","transID","type"]).reset_index(drop=True)
df_new
```

```python
df_new['start']=df_new.start.astype(int)
df_new['end']=df_new.end.astype(int)
df_new
```

```python
df_new.drop_duplicates(subset=gff3Cols,inplace=True,keep='first')
df_new.reset_index(drop=True,inplace=True)
```

```python
df_new[gff3Cols].to_csv("./hg38_p12_gencode29_refseq_filteredPseudogenes.gff3",index=False,sep='\t',header=False,quoting=csv.QUOTE_NONE)
```
